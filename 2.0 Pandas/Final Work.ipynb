```python
import pandas as pd
bike_df = pd.read_csv('data/citibike-tripdata.csv',sep=',')
bike_df_test = bike_df.copy()
bike_df_test.head()
```

```python
bike_df_test['age']=2018-bike_df_test['birth year']
bike_df_test.drop(['start station id','end station id','birth year'],axis=1)
bike_df_test[bike_df_test['age']>60].count()
```

```python
bike_df_test['starttime']=pd.to_datetime(bike_df_test['starttime'])
bike_df_test['stoptime']=pd.to_datetime(bike_df_test['stoptime'])
bike_df_test['trip duration']=bike_df_test['stoptime']-bike_df_test['starttime']
bike_df_test.head()
```

```python
bike_df_test['weekend']=bike_df_test['starttime'].dt.day_of_week
def get_weekend(weekday):
    if weekday == 5 or weekday == 6:
        weekday = 1
    else:
        weekday = 0
    return weekday
bike_df_test['weekend'] = bike_df_test['weekend'].apply(get_weekend)
bike_df_test['weekend'].value_counts()
```

```python
bike_df_test['time_of_day']=bike_df_test['starttime'].dt.hour
bike_df_test['time_of_day'].value_counts()
def get_time_type(hour) -> str:
    if hour>=0 and hour<=6:
        hour = 'night'
    elif hour>6 and hour<=12:
        hour = 'morning'
    elif hour>12 and hour<=18:
        hour = 'day'
    else:
        hour = 'evening'
    return hour
bike_df_test['time_of_day']=bike_df_test['time_of_day'].apply(get_time_type)
bike_df_test.drop(['time_of_day_pop'],axis=1)
bike_df_test['time_of_day'].value_counts()
143012/15085
```
