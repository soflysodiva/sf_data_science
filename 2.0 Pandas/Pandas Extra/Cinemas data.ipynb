```python
rating_mov = pd.read_csv('data/ratings_movies.csv')
rating_mov.head()
```

```python
#библиотека для регулярных выражений
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
rating_mov['year_release'] = rating_mov['title'].apply(get_year_release)
rating_mov.info()
```

```python
rating_mov[rating_mov['year_release']==2010].groupby('genres')['rating'].mean().sort_values()
```

```python
rating_mov.groupby('userId')['rating'].agg(
    ['count', 'mean']
).sort_values(['count', 'mean'], ascending=[True, False])
```

```python
topg = rating_mov[rating_mov['year_release']==2018].groupby('genres')['rating'].agg(
    ['count','mean'])
topg[topg['count']>10].sort_values(['mean'],ascending=False)
```

```python
rating_mov['date']=pd.to_datetime(rating_mov['date'])
rating_mov['year_rating']=rating_mov['date'].dt.year
topt = rating_mov.pivot_table(
    values='rating',
    index='year_rating',
    columns='genres',
    aggfunc=['mean'],
    fill_value=0
)
topt
```
